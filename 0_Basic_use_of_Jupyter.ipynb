## 一、Jupyter简介
### 1、基础介绍
是一个基于网页的用于交互计算的应用程序，支持Markdown语法。
- **【基于网页】**指的是Jupyter Notebook是以网页的形式打开，可以在网页页面中直接编写代码和运行代码。
- **【交互计算】**指的是，我们可以写一行代码运行一行代码。代码的运行结果也会直接在代码块下显示。方便及时查看代码结果，便于调试。
- **【Markdown语法】**Markdown是一种可以使用普通文本编辑器编写的标记语言，通过简单的标记语法，它可以使普通文本内容具有一定的格式。使用Markdown语法，可以将文本、图像、代码置于一个界面中，便于说明、调试。

### 2、Jupyter编写两种形式
Jupyter整体上有两种模式：编辑模式和命令模式。
- **命令模式**:主要对cell（文件中的代码或文本框）进行操作。例如：增、删、改、切换cell的模式等。
- **编辑模式**:编辑模式分，代码模式和markdown模式。编辑模式下有很多快捷键用于对文本或代码的格式进行操作。
