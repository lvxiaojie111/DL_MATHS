## windows下安装环境
### 一、安装anaconda
1、下载anaconda：https://www.continuum.io/downloads

2、单击下一步 进行安装 直到最后

3、有两个地方要选中：
    ①将anaconda加入path环境变量
    ②将anaconda注册为默认的python
### 二、启动命令提示符
1、点击左上角可以设置相关字体属性
### 三、利用anaconda建立虚拟环境
1、建立并切换到工作目录 md \pythonwork cd \pythonwork

2、建立anaconda虚拟环境 conda create --name tensorflow python=3.7 anaconda
    说明：加入anaconda建立虚拟环境时，会同时安装常用的数据分析包
    
3、启动虚拟环境：activate tensorflow

4、关闭虚拟环境：deactivate tensorflow

5、查看已有虚拟环境:conda-env list
### 四、虚拟环境中安装tensorflow与keras
1、启动虚拟环境：activate tensorflow

2、安装tensorflow:pip install tensorflow

3、安装keras:pip install keras
###  五、启动jupyter notebook
1、切换工作目录：cd \pythonwork

2、启动jupyter:jupyter notebook

3、建立文件，更改名字

4、shiftt+enter:执行后光标会移动到下一个单元格
   ctrl+enter:执行后，光标仍在当前的单元格
   
5、导入模块查看版本：import tensorflow as tf tf.__version__

6、导入模块查看版本：import keras keras.__version__

## linux下安装环境
### 一、复制下载anaconda的网址：anaconda forlinux 版本
### 二、下载这个文件：wget /  /  /.sh
### 三、安装anaconda
1、 终端输入:bash anaconda3-4.2.0-linux-x86_64.sh -b
   说明： 加入-b 是batch处理安装，会自动省略 阅读license条款，并自动安装到/home/hduser/anaconda3a下
   
2、编辑~/.bashrc加入模块路径：sudo gedit ~/.bashrc
   输入：export PATH="home/user/anaconda3/bin:$PATH"
   目的是为了将anaconda执行文件路径加入PATH，让我们在不同路径都可以执行anaconda.
   
3、重新注销登入，使得~/.bashrc修改生效
    source ~/.bashrc
    
4、在终端中查看python 版本：python --version

### 四、安装tf和keras
1、终端输入：pip install tensorflow

2、终端输入：pip isntall keras

### 五、启动jupyter notebook
安装anconda时会自动安装notebook,所以直接启动就可。

1、建立ipynotebook工作目录，终端中输入下列命令，切换到工作目录：

    mkdir -p ~/pywork 
    
    cd ~/pywork
    
    jupyter notebook
